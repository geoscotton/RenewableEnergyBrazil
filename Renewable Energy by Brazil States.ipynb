```python
import folium

import os 
cwd = os.getcwd()
cwd
# Load the shape of the zone (US states)

state_geo=cwd+'\\map.json'

# Load the data value of each state

state_data2 = pd.read_csv(cwd+"\\estados_renovavel.csv")
state_data = state_data2[['Estado','renovavel']]
state_data
```

```python

# Initialize the map:
m = folium.Map(location=[-15, -50], zoom_start=4)
 
# Add the color for the chloropleth:
m.choropleth(
 geo_data=state_geo,
 #name='choropleth',
 data=state_data,
 columns=['Estado', 'renovavel'],
 key_on='feature.id',
 fill_color='YlGn',
 fill_opacity=0.8,
 line_opacity=0.9,
 legend_name='% Renewable Energy Production in Brazil (2015)'
)
folium.LayerControl().add_to(m)
```

```python
#save html
m.save(cwd+"\\mapa.html")
```
